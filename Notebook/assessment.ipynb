## SCENARIO / WHAT IF ANALYSIS

1. If Q1 scores increase by 1 point for respondents under 25, how does the overall average
change?

```python
# Create a copy so we don’t overwrite original data
scenario_df = Surveyy_df.copy()

# Increase Q1 by +1 for respondents under age 25 (Under 18 and 18–24)
scenario_df.loc[scenario_df["Age Group"].isin(["Under 18", "18-24"]),
                "Q1: I have access to basic health services"] += 1
```

```python
# Recalculate averages
original_avg = Surveyy_df["Q1: I have access to basic health services"].mean()
new_avg = scenario_df["Q1: I have access to basic health services"].mean()

print("Original Average Q1:", round(original_avg, 2))
print("New Average Q1 After Scenario:", round(new_avg, 2))
print("Change:", round(new_avg - original_avg, 2))
```

After increasing Q1 scores for respondents aged under 25 (Under 18 and 18–24) by one point, the overall average Q1 score increased from 2.94 to 3.22, representing a change of 0.28.

Increasing Q1 scores by one point for respondents under 25 raised the overall average health access score from 2.94 to 3.22, a noticeable increase of 0.28. This indicates that improving access to basic health services for younger age groups has a meaningful positive impact on the overall population’s perceived access to healthcare.

2. Build a two-variable data table showing how Q2 and Q3 averages change with different regional weights.

```python
# Calculate the average Q2 (food security) and Q3 (school safety) scores for each region
regional_avgs = Surveyy_df.groupby("Region")[
    ['Q2: My household has enough food to eat',
    'Q3: My children can attend school safely']
].mean().reset_index()

regional_avgs
```

```python
weights = {
    "Equal": [1,1,1,1,1],                  # all equal
    "High Africa": [3,1,1,1,3],            # East + West Africa prioritized
    "High South Asia": [1,1,3,1,1],
    "High Middle East": [1,1,1,3,1]
}
```

```python
# Extract region names and their corresponding average Q2 and Q3 values into lists
regions = regional_avgs["Region"].tolist()
Avg_Q2 = regional_avgs["Q2: My household has enough food to eat"].tolist()
Avg_Q3 = regional_avgs["Q3: My children can attend school safely"].tolist()

results = []

# Loop through each weighting scenario to compute weighted averages

for scenario, w in weights.items():
    total_weight = sum(w)

    # Compute weighted averages for Q2 and Q3 based on selected weight pattern

    weighted_Q2 = np.average(Avg_Q2, weights=w)
    weighted_Q3 = np.average(Avg_Q3, weights=w)
    results.append([scenario, round(weighted_Q2, 2), round(weighted_Q3, 2)])


# Convert results into a DataFrame for easy viewing
data_table = pd.DataFrame(results, columns=["Scenario", "Weighted Avg Q2", "Weighted Avg Q3"])

# Display final weighted scenario comparison table
print(data_table)
```

When all regions are weighted equally, the overall food security (Q2) and school safety (Q3) averages are 2.74 and 2.96, respectively. Prioritizing East and West Africa (High Africa scenario) lowers the overall averages to 2.66 for Q2 and 2.87 for Q3. This occurs because these African regions have comparatively lower scores, so giving them more weight pulls down the global averages. Conversely, prioritizing regions with higher scores, such as the Middle East (High Middle East scenario), increases the overall Q3 average to 3.09. These results illustrate that regional disparities significantly influence overall outcomes, and targeted weighting can reveal which regions drive the global metrics up or down.
